```python
import pandas as pd
import numpy as np
```

```python
file = "/users/saurin/desktop/Resources/purchase_data.csv"
```

```python
df = pd.read_csv(file)
df.head()
```

```python
df_new = df.drop_duplicates(subset="SN")
df_new.head()
```

```python
# total_players 
total_players = df_new["SN"].count()
total_players = [{"Total Players":total_players}]
df_2 = pd.DataFrame(total_players)
df_2
```

```python
# Purchasing Analysis

unique_items = len(df["Item Name"].drop_duplicates())
average = df["Price"].mean()
total_items = df["Item Name"].count()
revenue = df["Price"].sum()

purchasing_analysis = [{"Number Of Unique Items":unique_items, "Average Purchase Price":average, "Total Number of Purchases":total_items, "Total Revenue":revenue}]
df_analysis = pd.DataFrame(purchasing_analysis)
df_analysis
```

```python
# Gender Demographics

gender_group_1  = df.groupby(["Gender"])
gender_group = df_new.groupby(["Gender"])
gender_group["SN"].count()

df_4 = pd.DataFrame({"Total Players":gender_group["Gender"].count()})
df_4["Percentage Of Players"]= ((gender_group["Gender"].count()/ df_new["Gender"].count())*100).map("{:.2f}%".format)
df_4
```

```python
count_df = gender_group_1["Item Name"].count()
count_df
```

```python
count = gender_group_1["Item Name"].count()
avg_pur_price = (gender_group_1["Price"].mean()).map("${:.2f}".format)
tot_value = gender_group_1["Price"].sum()
tot_players = gender_group["Gender"].count()
avg = tot_value/tot_players
purchasing_analysis_ = pd.DataFrame({"Purchase Count":count, "Average Purchase Price":avg_pur_price, "Total Purchase Value":(tot_value).map("${:.2f}".format), "Avg Total Purchase Per Person":(avg).map("${:.2f}".format)})
purchasing_analysis_gender = pd.DataFrame(purchasing_analysis_)
purchasing_analysis_gender
```

```python
# Age Demographics
bins = [0, 10, 15, 20, 25, 30, 35, 40, 45]
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
df_new["Total_count"] = pd.cut(df_new["Age"], bins, labels=group_labels)
df_new.head()
```

```python
grouped_age = df_new.groupby(["Total_count"])
count_age_df = grouped_age["Age"].count()/df_new["SN"].count()
count_age_df
df_age_group = pd.DataFrame({"Total Count":grouped_age["Age"].count(),"Percentage of Players":((grouped_age["Age"].count()/df_new["SN"].count())*100).map("{:.2f}%".format)})
df_age_group
                            
```

```python
df["Total_count"] = pd.cut(df["Age"],bins, labels=group_labels)
df.head()
```

```python
grouped_age_1 = df.groupby("Total_count")
count = grouped_age_1["Item Name"].count()
avg_pur_price = (grouped_age_1["Price"].mean()).map("${:.2f}".format)
tot_value = grouped_age_1["Price"].sum()
tot_players = grouped_age_1["Gender"].count()
avg = (tot_value/tot_players).map("${:.2f}".format)
purchasing_analysis_ = pd.DataFrame({"Purchase Count":count, "Average Purchase Price":avg_pur_price, "Total Purchase Value":(tot_value).map("${:.2f}".format), "Avg Total Purchase Per Person":avg})
purchasing_analysis_gender = pd.DataFrame(purchasing_analysis_)
purchasing_analysis_gender
```

```python
# Top Spenders
spenders = df.groupby("SN")
total = (spenders["Price"].sum()).sort_values(ascending=False)
total.head()
```

```python
purchase_count = (spenders["Price"].count()).sort_values(ascending=False)
purchase_count.head(10)
```

```python
average = total/purchase_count
average_1 = average.sort_values(ascending=False)
average_1
```

```python
df_5 = pd.DataFrame({"Purchase Count":purchase_count, "Purchase Value":total,"Average Purchase Price":average_1})
df_5
top_spenders_df = df_5.sort_values("Purchase Value",ascending=False)
top_spenders_df.head()
```

```python
item_df = df[["Item ID", "Item Name", "Price"]]
item_df
```

```python
grouped_items = item_df.groupby(["Item ID","Item Name"])
sorted_ = grouped_items.count()
sorted_items_df = sorted_.sort_values("Price", ascending=False)
sorted_items_df
```

```python
total_value = grouped_items.sum()
total_value_df = total_value.sort_values("Price", ascending=False)
total_value_df
```

```python
merged_df = pd.merge(sorted_items_df, total_value_df, on=["Item ID","Item Name"])
merged_df
```

```python
rename_df = merged_df.rename(columns={"Price_x":"Purchase Count", "Price_y":"Total Purchase Value"})
rename_df.head()
```

```python
rename_df["Item Price"] = rename_df["Total Purchase Value"]/rename_df["Purchase Count"]
rename_df.head()
```

```python
# Most profitable items
most_profitable_df = rename_df.sort_values("Total Purchase Value", ascending=False)
most_profitable_df.head()
```
